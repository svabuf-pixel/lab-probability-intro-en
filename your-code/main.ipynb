## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
# YOUR CODE HERE
white_balls = 10
red_balls = 12
black_balls = 8
total_balls = white_balls + red_balls + black_balls

# 1)
taking_white_ball = white_balls / total_balls
# 2)
taking_white_black_without_replacing = taking_white_ball * (black_balls / (total_balls - 1))
# 3)
taking_white_black_replacing_replacing = taking_white_ball * (black_balls / total_balls)
# 4)
taking3_balls_different_colours = 6 * (
    taking_white_ball * (red_balls / (total_balls - 1)) * (black_balls / (total_balls - 2))
)

print(
    f"Ex1: {taking_white_ball:.2}\nEx2: {taking_white_black_without_replacing:.2}\nEx3: {taking_white_black_replacing_replacing:.2}\nEx4: {taking3_balls_different_colours:.2}"
)
```

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

```python
# YOUR CODE HERE
cloudy = 0.4
not_cloudy = 1 - cloudy
rainy_when_is_cloudy = 0.5
rain_when_is_not_cloudy = 0.2

rain = (rainy_when_is_cloudy * cloudy) + (rain_when_is_not_cloudy * not_cloudy)
print(f"The chance to reain is: {rain}%")
```

## Challenge 3

A and B are events of a probability such that $P(A) = 0.3$, $P(B) = 0.6$ and $P(A \cap B) = 0.1$

Which of the following statements are true?  
**HINT: draw a big Venn Diagram by your side**  
* $P(\overline{A}) = 0.7$
* $P(A \cup B) = 0.6$
* $P(A \cap \overline{B}) = 0.2$
* $P(A \cup \overline{A}) = 1$  
* $P(A \cap (B \cup \overline{B})) = 0.4$
* $P(\overline{A} \cap \overline{B}) = 0.3$
* $P(\overline{A \cap B}) = 0.9$

```python
# YOUR CODE HERE
prob_A = 0.3
prob_B = 0.6
A_and_B = 0.1
complement_of_prob_A = 1 - prob_A
complement_of_prob_B = 1 - prob_B
A_or_B = prob_A + prob_B - A_and_B
A_happening_and_B_not = prob_A - A_and_B
A_happ_or_A_not_happ = prob_A + complement_of_prob_A
A_and_B_or_not_B = prob_A + (1 - (complement_of_prob_B + prob_B))
not_a_and_not_b = 1 - A_or_B
# The probability that it is not the case that both happen at once. (This is a De Morgan's Law application!)
last = 1 - A_and_B

print(
    f"1: {complement_of_prob_A}\n2: {A_or_B}\n3: {A_happening_and_B_not}\n4: {A_happ_or_A_not_happ}\n5: {A_and_B_or_not_B}\n6: {not_a_and_not_b}\n7: {last}"
)
```

## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

## BONUS

You roll 2 dice. What is the probability that the sum of the numbers is 7? And 8? And 9? Then simulate 10000 times and draw a countplot.

your explanation here

```python
# YOUR CODE HEREimport numpy as np
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Simulate 10,000 rolls for two dice
# np.random.randint(1, 7, size) generates random integers from 1 to 6
dice_1 = np.random.randint(1, 7, 10000)
dice_2 = np.random.randint(1, 7, 10000)

# 2. Calculate the sum of the two dice for each roll
sums = dice_1 + dice_2

# 3. Create the countplot to visualize the distribution
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
plot = sns.countplot(x=sums, palette="viridis")

# Add labels and title
plt.title("Simulation of 10,000 Dice Rolls: Sum of 2 Dice", fontsize=15)
plt.xlabel("Sum of the two dice", fontsize=12)
plt.ylabel("Frequency (Number of times)", fontsize=12)

# Save the plot
plt.savefig("dice_simulation.png")

# 4. Calculate and print the simulated probabilities for 7, 8, and 9
unique, counts = np.unique(sums, return_counts=True)
probabilities = dict(zip(unique, counts / 10000))

print(f"Simulated Probability of Sum 7: {probabilities.get(7, 0):.4f}")
print(f"Simulated Probability of Sum 8: {probabilities.get(8, 0):.4f}")
print(f"Simulated Probability of Sum 9: {probabilities.get(9, 0):.4f}")
```
